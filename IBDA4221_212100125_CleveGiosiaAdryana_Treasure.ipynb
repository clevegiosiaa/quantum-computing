```python
from qiskit import *
```

# Original Circuit

```python
qc = QuantumCircuit(3)

qc.h(0)
qc.cx(0,1)
qc.h(2)

qc.barrier(0,1,2)

qc.cx(2,1)
qc.x(2)
qc.cx(2,0)
qc.x(2)

qc.barrier(0,1,2)

qc.swap(0, 1)
qc.x(0)
qc.x(1)
qc.cx(2,1)
qc.x(2)
qc.cx(2,0)
qc.x(2)

qc.draw(output='mpl')
```

```python
from qiskit import Aer, execute
measured_qc = qc.measure_all(inplace=False)
simulator = Aer.get_backend('qasm_simulator')
counts = execute(measured_qc, backend=simulator, shots=1000).result().get_counts(measured_qc)
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

```python
# Membuat sirkuit kuantum baru
qc_simplified = QuantumCircuit(3)

# Menambahkan gate yang tersisa setelah penyederhanaan
qc_simplified.h(0)
qc_simplified.cx(0,1)
qc_simplified.h(2)

# Melakukan pengukuran
qc_simplified.measure_all()

# Menggambar sirkuit
qc_simplified.draw(output='mpl')
```

```python
from qiskit import Aer, execute
simulator = Aer.get_backend('qasm_simulator')
counts = execute(qc_simplified, backend=simulator, shots=1000).result().get_counts(qc_simplified)
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```
